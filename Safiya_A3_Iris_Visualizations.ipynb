```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.heatmap(iris.corr(), annot = True)
```

```python
sns.lineplot(x="petal_length", y="petal_width", data = iris, ci = None, color = 'purple')
```

```python
fig=plt.figure(figsize = (15,15))
ax = fig.add_subplot(2,1,1)
sns.scatterplot(x="sepal_length", y="petal_length", data = iris,color = 'orange')
ax1 = fig.add_subplot(2,1,2)
sns.scatterplot(x="sepal_length", y="petal_width", data = iris,color = 'magenta')
```

```python
sns.countplot(x="species", data = iris)
```

```python
fig=plt.figure(figsize = (10,15))
ax1 = fig.add_subplot(4,1,1)
sns.distplot(iris["petal_length"], color = 'purple')

ax2 = fig.add_subplot(4,1,2)
sns.distplot(iris["petal_width"], color = 'pink')

ax3 = fig.add_subplot(4,1,3)
sns.distplot(iris["sepal_length"], color = 'magenta')

ax4 = fig.add_subplot(4,1,4)
sns.distplot(iris["sepal_width"], color = 'red')
```
