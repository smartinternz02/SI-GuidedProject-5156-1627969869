```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([("one",OneHotEncoder(),[1,2,3,4,6,7,8,10,15])],remainder = "passthrough")
x = ct.fit_transform(x)
```

```python
ct1 = ColumnTransformer([('one', OneHotEncoder(), [0])], remainder = 'passthrough')
y= ct1.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential 

from tensorflow.keras.layers import Dense
```

```python
categ = Sequential()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#Input Layer
categ.add(Dense(units = 51, kernel_initializer = "random_uniform", activation = "relu"))
```

```python
#Hidden Layer1
categ.add(Dense(units = 102, kernel_initializer = "random_uniform", activation = "relu"))
```

```python
#Hidden Layer2
categ.add(Dense(units = 102, kernel_initializer = "random_uniform", activation = "relu"))
```

```python
#Output Layer
categ.add(Dense(units = 2, kernel_initializer = "random_uniform", activation = "softmax"))
```

```python
categ.compile(optimizer = "rmsprop", loss = "categorical_crossentropy", metrics = ["accuracy"]) 
```

```python
categ.fit(x_train, y_train, epochs = 500)
```

```python
y_predict = categ.predict(x_test)
```
